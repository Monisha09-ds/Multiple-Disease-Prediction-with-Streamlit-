```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Parkinsson disease.csv')
df.head()
```

```python
X = df.drop(columns=['name','status'],axis=1)
y=df['status']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler()
X_scaled = scaled.fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape,'\n',Y_train.shape,'\n',X_test.shape,'\n',Y_test.shape)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(train_prediction,Y_train)
print("Accuracy on testing data : ",train_data_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(test_prediction,Y_test)
print("Accuracy on testing data : ",test_data_accuracy)
```

```python
def parkinson_prediction(inp):
    inp_array = np.asarray(inp)
    input_data_reshaped = inp_array.reshape(1,-1)
    prediction = model.predict(input_data_reshaped)
    if (prediction[0]== 0):
        print('The Person has Parkinson')
    else:
        print('The Person does not have Parkinson')
```
